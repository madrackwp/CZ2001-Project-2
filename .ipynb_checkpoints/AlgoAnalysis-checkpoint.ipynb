```python
import random
import time
from statistics import *
```

```python
def generateHospitalList(h,noOfNodes): #h refers to the no. of hospital nodes
    hospitalNodes =[]
    while(h>0):
        node = random.randrange(1, noOfNodes)
        while (node in hospitalNodes):
            node = random.randrange(1, noOfNodes)
        hospitalNodes.append(node)
        h-=1
    return hospitalNodes
```

### Varying k while keeping h constant at h=6

```python
def test(adjacencyList, k, h,opPath):
    timelist = []
    for x in range(1,101):
        wipeTextFile(opPath)
        hospitals = generateHospitalList(h,1000)
        tic = time.time()
        for startingNode in adjacencyList:
            hospitalList = hospitals.copy()
            BFS(adjacencyList, startingNode, k, hospitalList, opPath)
        toc = time.time()
        timelist.append(toc-tic)

    print("Mean: ",mean(timelist))
    print("Median: ",median(timelist))
    print("SD: ",stdev(timelist))
```

### Analysis:
As we increase the density of the sparse graph, we can observe a decrease in the time taken to complete the algorithm. One possible conclusion that we came up with was the increase in density has allowed for shorter routes to hospitals to appear. Hence this will decrease the time taken for the graph to find a path to the nearest hospital.

### Observation:
Increasing the density of the graph closer to a full graph increases the time taken for the algorithm to complete

### Analysis:
Our previous analysis on the time complexity depends quite on V and not E as we assumed it to be a sparse graph. However as we increase the density of the graph towards a more full one, the difference between V and E decreases which means that we will have to start taking E into consideration when it comes to time complexity analysis. Hence this observation does still support our previous analysis.

At the same time, we can observe that the efficiency of the adjacency list will start to diminish. The benefit of the adjacency list is more apparent when the graph is sparse. However, as the graph becomes more and more dense, each node is connected to more nodes, increasing the list linked to each parent node. Thus the access time for the entire algorith will come closer to O(|V|^3)
